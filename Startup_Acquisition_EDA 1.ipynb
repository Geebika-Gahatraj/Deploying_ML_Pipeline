###**UNIVARIATE ANALYSIS**

```python
plt.figure(figsize=(10,6))
sns.histplot(cleaned_df.founded_at, bins = 10)
```

```python
plt.figure(figsize=(10,6))
sns.histplot(cleaned_df.active_days, bins = 10)
```

```python
plt.figure(figsize=(10,6))
sns.histplot(cleaned_df.first_milestone_at, bins = 10)
```

```python
plt.figure(figsize=(10,6))
sns.histplot(cleaned_df.last_milestone_at, bins = 10)
```

###**BIVARIATE ANALYSIS**


```python
# for categorical datas
# Bar graph
plt.figure(figsize=(26,12))
sns.countplot(x='category_code', data=cleaned_df, order= cleaned_df.category_code.value_counts().index, palette='Accent')
plt.xlabel('category_code')
plt.ylabel('count')
```

```python
plt.figure(figsize=(26,16))
sns.countplot(x='country_code', data=cleaned_df, order= cleaned_df.country_code.value_counts().index, palette='pastel')
plt.xlabel('country_code')
plt.ylabel('count')
```

```python
plt.figure(figsize=(10,4))
sns.countplot(x='status' , data=cleaned_df, palette='viridis')
plt.xlabel('status', fontsize=14)
plt.show()
```

```python
sns.scatterplot(x=cleaned_df['active_days'], y=cleaned_df['category_code'], hue=cleaned_df['status'])
```

```python
cleaned_df.head()
```

```python
sns.lmplot(x="first_funding_at", y="active_days", hue="status", data=cleaned_df)
```

```python
sns.lmplot(x="founded_at", y="active_days", hue="status", data=cleaned_df)
```

```python
# pairplot to show the relation of each feature with every other feature
plt.figure(figsize=(15,15))
sns.pairplot(cleaned_df)
```
